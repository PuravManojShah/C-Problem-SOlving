#### Queue Using Array

```c++
//
//  main.cpp
//  QueueUsingArray
//
//  Created by PURAV SHAH on 6/8/20.
//  Copyright © 2020 PURAV SHAH. All rights reserved.
//

#include <iostream>
using namespace std;
template <typename T>
class QueueUsingArray{
    T *data;
    int nextIndex;
    int capacity;
public:
    QueueUsingArray(){
        capacity=5;
        data=new T[capacity];
        nextIndex=0;
    }
    int size(){
        return this->nextIndex;
    }
    bool isEmpty(){
        return this->nextIndex==0;
    }
    T front(){
        if (this->nextIndex==0){
            cout<<"Queue is empty"<<endl;
            return 0;
        }
        return this->data[0];
    }
    void increaseSize(){
        this->capacity=2*this->capacity;
        T * newData=new T[this->capacity];
        for (int i=0;i<nextIndex;i+=1){
            newData[i]=this->data[i];
        }
        delete this->data;
        this->data=newData;
    }
    void enqueue(int element){
        if (nextIndex==capacity){
            increaseSize();
        }
        data[this->nextIndex]=element;
        this->nextIndex+=1;
        return;
    }
    T dequeue(){
        if (this->nextIndex==0){
            cout<<"QUEUE IS EMPTY!"<<endl;
            return 0;
        }
        int element=this->data[0];
        for (int i=1;i<nextIndex;i+=1){
            this->data[i-1]=this->data[i];
        }
        this->nextIndex-=1;
        return element;
    }
    
};
int main() {
    QueueUsingArray<int> q;
    q.enqueue(10);
    q.enqueue(20);
    q.enqueue(30);
    cout<<q.dequeue()<<endl;
    q.enqueue(50);
    cout<<q.dequeue()<<endl;
    cout<<q.dequeue()<<endl;
    cout<<q.dequeue()<<endl;
    cout<<q.dequeue()<<endl;
    cout<<q.isEmpty()<<endl;
}
```

```c++
//
//  main.cpp
//  StackUsingLinkedList
//
//  Created by PURAV SHAH on 6/8/20.
//  Copyright © 2020 PURAV SHAH. All rights reserved.
//

#include <iostream>
using namespace std;
template <typename T>
class Node {
    public :
    T data;
    Node<T> *next;
    
    Node(T data) {
        this -> data = data;
        next = NULL;
    }
};

template <typename T>
class Stack {
    Node<T> *head;
    int size;        // number of elements prsent in stack
    
    public :
    
    Stack() {
        head=NULL;
        size=0;
    }
    
    int getSize() {
        return size;
    }
    
    bool isEmpty() {
        return size==0;
    }
    
    void push(T element) {
         Node<T> * newNode= new Node<T>(element);
        if(head == NULL){
            head=newNode;
            size+=1;
            return;
        }
        Node<T> * temp=head;
        while(temp->next != NULL){
            temp=temp->next;
        }
        temp->next=newNode;
        this->size+=1;
        return;
    }
    
    T pop() {
        // Return 0 if stack is empty. Don't display any other message
        if (size==0){
            return 0;
        }
        Node<T> * temp=head;
        Node<T>* prev=NULL;
        while(temp->next != NULL){
            prev=temp;
            temp=temp->next;
        }
        prev->next=NULL;
        size-=1;
        return temp->data;
    }
    
    T top() {
        // Return 0 if stack is empty. Don't display any other message
        if (size==0){
            return 0;
        }
        Node<T> * temp=head;
        while(temp->next != NULL){
            temp=temp->next;
        }
        return temp->data;
    }
    
  
    
};


int main() {
    Stack<int> s;
    s.push(10);
    s.push(20);
    cout<<s.pop()<<endl;
   
}
```
